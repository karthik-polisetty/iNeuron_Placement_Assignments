### Get all Pokemons that have less than 4 weaknesses

```python
import ast
df['weaknesses']=df['weaknesses'].apply(ast.literal_eval)
```

### Get all Pokemons that do not have more than 2 evolutions

```python
def evolution(i):
    if i==0:
        return 0
    else:
        i = ast.literal_eval(i)
        return len(i)
```

```python
df['prev_evolution'].fillna(0,inplace=True)
df['next_evolution'].fillna(0,inplace=True)
```

```python
# created a new column with evolution count
df['evolution_count']= df['prev_evolution'].apply(evolution) + df['next_evolution'].apply(evolution)
```

```python
df[df['evolution_count']<=2]
```

#### Get all Pokemons whose spawn time is less than 300 seconds. Note - spawn time format is "05:32”, so assume “minute: second” format and perform the analysis.

```python
def convert_time(i):
    mn = int(i[0:2])
    sec = int(i[3:5])
    val = mn*60 + sec
    return val
```

```python
# filled null values with "00:00"
df['spawn_time'].fillna("00:00",inplace=True)
```

```python
# converted the spawn_time into seconds
df["spawn_time"] = df['spawn_time'].apply(convert_time)
```

```python
# pokemons with spawn time less than 300
df[df['spawn_time']<300]
```
